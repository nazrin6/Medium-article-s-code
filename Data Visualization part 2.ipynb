```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

### Heatmap

```python
df = pd.DataFrame(np.random.random((5,5)), columns=["a","b","c","d","e"])

# Default heatmap
p1 = sns.heatmap(df)
```

### Jointplot 

```python
x, y = np.random.RandomState(8).multivariate_normal([0, 0], [(1, 0), (0, 1)], 1000).T
df = pd.DataFrame({"x":x,"y":y})
p = sns.jointplot(data=df,x='x', y='y')
```

### pairplot

```python
df = sns.load_dataset('iris')
plt.rcParams['figure.figsize']=1,1
sns.pairplot(df, hue ='species')
plt.show()
```

### Count plot

```python
plt.rcParams['figure.figsize']=5,5
sns.countplot(data = df,x = 'species')
```

### Strip plot

```python
plt.rcParams['figure.figsize']=5,5
sns.stripplot(y = df['species'], x = df['sepal_length'])
```

### Swarm plot

```python
sns.swarmplot(y = df['species'], x = df['sepal_length'])
```

### density plot

```python
sns.distplot(df['sepal_length'], hist=False, kde=True, 
             bins=int(180/5), color = 'darkblue', 
             kde_kws={'linewidth': 4})
```

### KDE plot

```python
sns.kdeplot(df.loc[(df['species']=='virginica'),
            'sepal_length'], color='b', shade=True, Label='Iris_Virginica')
```

```python
sns.kdeplot(df.loc[(df['species']=='setosa'),
            'sepal_length'], color='r', shade=True, Label='Setosa')
  
sns.kdeplot(df.loc[(df['species']=='virginica'), 
            'sepal_length'], color='b', shade=True, Label='Virginica')
```

```python
iris_setosa = df.query("species=='setosa'")
iris_virginica = df.query("species=='virginica'")
```

```python
sns.kdeplot(iris_setosa['sepal_length'], 
            iris_setosa['sepal_width'],
            color='r', shade=True, Label='Iris_Setosa',
            cmap="Reds", shade_lowest=False)
```

### Lm plot

```python
sns.lmplot(x= 'sepal_length', y = 'petal_length', data = df, hue = 'species')
```

### Cat plot

```python
sns.catplot(data = df, x = 'sepal_length', y = 'species', kind = 'strip')
```

```python
sns.catplot(data = df, x = 'sepal_length', y = 'species', kind = 'bar')
```
